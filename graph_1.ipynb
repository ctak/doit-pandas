```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
overwatch = pd.read_csv('./data/overwatch.csv')
lol = pd.read_csv('./data/lol.csv')
print(overwatch.head())
print(lol.head())
overwatch.index = overwatch['date']
lol.index = lol['date']
print(overwatch.head())
print(lol.head())
```

```python
overwatch.set_index('date')
lol.set_index('date')
# Creating a plot with time as x-axis and search on y-axis
fig, ax = plt.subplots(figsize=(20, 8)) # inch 라는 게 너무 미국적이네.
ax.plot(overwatch.index.values, overwatch['search'])
ax.plot(lol.index.values, lol['search'])
ax.set_xlabel('Date')
ax.set_ylabel('Number of Searches')
ax.legend(['Overwatch', 'LoL'])
plt.show()

fig.savefig('overwatch_vs_lol.png', dpi=600)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(20, 8))
ax[0].plot(overwatch.index.values, overwatch['search'], linestyle='--', marker='o', color='b')
ax[0].set_title('Overwarch Search Record')
ax[1].plot(lol.index.values, lol['search'], linestyle = '--', marker = 'v', color = 'r')
ax[1].set_title('LoL Search Record')
fig.tight_layout()
plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10), marker='.', markersize=5, linestyle='--')
plt.plot(np.arange(10), np.arange(10)**2, marker='o', markersize=5, linestyle='-.')
plt.plot(np.arange(10), np.log(np.arange(10)), marker='+', markersize=15, linestyle=':', color='g', alpha=0.5)

plt.title('test.', fontsize=10)

plt.xlabel('x축', fontsize=10)
plt.ylabel('y축', fontsize=10)

plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.legend(['10', '10**2', 'log'], fontsize=20)

plt.xlim(0, 9)
plt.ylim(0.5, 10)

plt.show()
```

```python
import numpy as np
a = np.random.rand(5)
print(a)
b = np.random.rand(2, 3)
print(b)
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 3)
plt.rcParams['font.size'] = 12

a = np.random.rand(1000)
b = np.random.rand(10000)
c = np.random.rand(100000)

plt.hist(a, bins=100, density=True, alpha=0.5, label='n=1000', histtype='step')
plt.hist(b, bins=100, density=True, alpha=0.75, label='n=10000', histtype='step')
plt.hist(c, bins=100, density=True, alpha=1.0, label='n=100000', histtype='step')

plt.ylim(0, 2.5)
plt.legend()
plt.show() # 샘플의 개수가 1000, 10000, 100000개로 증가할수록 더욱 균일한 분포를 보임을 알 수 있습니다.
```

```python
a = np.random.randint(2, size=5) # (0, 2) 범위에서 다석개의 임의의 정수. 0 이상 2미만.
print(a)

b = np.random.randint(2, 4, size=5)
print(b)

c = np.random.randint(1, 5, size=(2, 3))
print(c)
```

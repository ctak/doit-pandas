```python
avg_life_exp_by_year = df.groupby('year').lifeExp.mean()
print(avg_life_exp_by_year)
```

```python
# agg 메서드로 사용자 함수와 groupby 메서드 조합하기
# 여기서 values 는 groupby 된 year DataFrame 에 lifeExp 리스트네.
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum/n

agg_my_mean = df.groupby('year').lifeExp.agg(my_mean)
print(agg_my_mean)
```

```python
# 2개의 인자값을 받아 처리하는 사용자 함수와 groupby 메서드
# diff_value 는 naming parameter 로 받는군.
def my_mean_diff(values, diff_value):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean = sum / n
    return mean - diff_value

# 전체 평균 수명
global_mean = df.lifeExp.mean()
print(global_mean)

agg_mean_diff = df.groupby('year').lifeExp.agg(my_mean_diff, diff_value=global_mean)
print(agg_mean_diff)
print(type(agg_mean_diff))
```

```python
# 이런 식으로 배열로 함수를 던져주네.
import numpy as np
gdf = df.groupby('year').lifeExp.agg([np.count_nonzero, np.mean, np.std])
print(gdf)
```

```python
gdf_dic = df.groupby('year').agg({ 'lifeExp': 'mean', 'pop': 'median', 'gdpPercap': 'median' })
print(gdf_dic)
```

```python
# 평균과 표준편차의 차이를 표준점수 라고 한다.
def my_zscore(x):
    return (x - x.mean()) / x.std()

df['transform_z'] = df.groupby('year').lifeExp.transform(my_zscore)

print(df)
df[df.transform_z > 1]
```

```python
import seaborn as sns
import numpy as np

np.random.seed(42)

tips_10 = sns.load_dataset('tips').sample(10)
print(tips_10)
tips_10.loc[np.random.permutation(tips_10.index)[:4], 'total_bill'] = np.NaN
print(tips_10)
```

```python
# 누락값을 단순히 평균값으로 채우면 안된다.
# 여성과 남성으로 구분하여 채워주자.

count_sex = tips_10.groupby('sex').count()
print(count_sex)
```

```python
def fill_na_mean(x):
    avg = x.mean()
    return x.fillna(avg)

total_bill_group_mean = tips_10.groupby('sex').total_bill.transform(fill_na_mean)
print(total_bill_group_mean)
tips_10['fill_total_bill'] = total_bill_group_mean
print(tips_10)
```
